```python
# Настройка стиля графиков
sns.set(style="whitegrid")

# Построение гистограмм для распределения 'Mean Radius', 'Radius Std.Error' и 'Worst Radius'
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.histplot(data=df, x='Mean Radius', kde=True, hue='Diagnosis', element="step", stat="density", common_norm=False)
plt.title('Распределение Mean Radius')

plt.subplot(1, 3, 2)
sns.histplot(data=df, x='Radius Std.Error', kde=True, hue='Diagnosis', element="step", stat="density", common_norm=False)
plt.title('Распределение Radius Std.Error')

plt.subplot(1, 3, 3)
sns.histplot(data=df, x='Worst Radius', kde=True, hue='Diagnosis', element="step", stat="density", common_norm=False)
plt.title('Распределение Worst Radius')

plt.tight_layout()
plt.show()
```

```python
# Построение boxplot для параметров 'Mean Radius', 'Radius Error' и 'Worst Radius'
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.boxplot(data=df, x='Diagnosis', y='Mean Radius')
plt.title('Boxplot для Mean Radius')

plt.subplot(1, 3, 2)
sns.boxplot(data=df, x='Diagnosis', y='Radius Std.Error')
plt.title('Boxplot для Radius Std.Error')

plt.subplot(1, 3, 3)
sns.boxplot(data=df, x='Diagnosis', y='Worst Radius')
plt.title('Boxplot для Worst Radius')

plt.tight_layout()
plt.show()
```

```python
# Выбор интересующих столбцов для парной диаграммы
selected_columns = ['Mean Radius', 'Radius Std.Error', 'Worst Radius', 'Diagnosis']

# Создание парной диаграммы
sns.pairplot(df[selected_columns], hue='Diagnosis', markers=["o", "s"])
plt.suptitle('Парные диаграммы для выбранных параметров', y=1.02)
plt.show()
```

```python
newDf = df
newDf['Diagnosis'] = df['Diagnosis'].replace({'Benign': 0, 'Malignant': 1})
# Вычисление корреляционной матрицы
corr_matrix = newDf[['Mean Radius','Radius Std.Error','Worst Radius']].corr()

# Визуализация корреляционной матрицы
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm',square=True)
plt.title('Корреляционная матрица радиуса')
plt.show()
```

```python
# Сводка среднего значения и стандартного отклонения параметров для злокачественных и доброкачественных опухолей
summary = df.groupby('Diagnosis').agg(['mean', 'std']).T
print(summary)
```
